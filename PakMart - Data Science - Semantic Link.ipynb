## PakMart Traders – Semantic Link

Uses `sempy.fabric` to explore relationships in the `pakmart_gold` Power BI dataset and detect functional dependencies.

```python
%pip install semantic-link
```

```python
import sempy.fabric as fabric

df_datasets = fabric.list_datasets()
df_datasets
```

### Dataset relationships – pakmart_gold star schema

```python
dataset = 'pakmart_gold'
relationships = fabric.list_relationships(dataset)
relationships
```

Expected relationships:
- `fact_sale.CityKey`        → `dimension_city.CityKey`
- `fact_sale.StockItemKey`   → `dimension_stock_item.StockItemKey`
- `fact_sale.InvoiceDateKey` → `dimension_date.Date`
- `fact_sale.SalespersonKey` → `dimension_employee.EmployeeKey`
- `fact_sale.CustomerKey`    → `dimension_customer.CustomerKey`

```python
from sempy.relationships import plot_relationship_metadata
plot_relationship_metadata(relationships)
```

### Functional dependencies in dimension_employee

```python
employee = fabric.read_table(dataset, 'dimension_employee')
employee.head(10)
```

```python
dependencies = employee.find_dependencies()
dependencies
```

```python
from sempy.dependencies import plot_dependency_metadata
plot_dependency_metadata(dependencies)
```

### Functional dependencies in dimension_city (provinces & territories)

```python
city = fabric.read_table(dataset, 'dimension_city')
city_deps = city.find_dependencies()
plot_dependency_metadata(city_deps)
```
